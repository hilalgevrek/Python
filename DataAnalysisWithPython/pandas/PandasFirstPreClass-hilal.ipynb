```python
pd.value_counts(titanic_df["Survived"]).plot.bar()  # Hayatta kalanların sayısı ile grafik oluşturmasını istedik. 0 ölenler, 1 yaşayanlar şeklinde.
```

```python
titanic_df["Survived"].mean()  # Hayatta kalma ortalamasını bulduk. 100 kişiden 38 kadarı hayatta kalmış.
```

```python
titanic_df.groupby(["Sex"]).mean()  # Cinsiyete göre ortalamalar
```

```python
titanic_df.groupby(["Sex", "Pclass"]).mean()  # First class ve female olanların hayatta kalma oranı %96
```

```python
titanic_df[titanic_df["Age"]<18].groupby(["Sex", "Pclass"]).mean()
```
