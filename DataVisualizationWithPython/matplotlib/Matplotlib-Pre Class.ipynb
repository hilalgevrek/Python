```python
x = np.linspace(0,10,10)  # Veriyi 0'dan 10'a kadar 10 nokta halinde döndürür
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()
plt.subplot(2,2,1)  # (row,column)
plt.plot(x,y1,"ro")
plt.title("$y_1=x$")

plt.subplot(2,2,2)
plt.plot(x,y1,"g--")
plt.title("$y_2=x^2$")  # Dolar işaretleri arasına _ şeklinde belirtilen değerlerin üs ve alta yazılmasını sağladık

plt.subplot(2,2,3)
plt.plot(x,y1,"b^")
plt.title("$y_3=x^3$")

plt.subplot(2,2,4)
plt.plot(x,y1,"ks")
plt.title("$y_4=\sqrt{x}$")
```

```python
style.use("ggplot")
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 129]
plt.hist(blood_sugar)  # histogram oluşturduk  # default bin sayısı 10'dur (bin ne işe yarıyor tam anlamadım. x axis'deki değişim noktalarını gösteriyor)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95, color = "#038694")  # color="g" gibi de yazabilirdim.
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Bloos sugar analysis")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,120,150], rwidth=0.95, color=["green","orange"], label=["men", "women"], orientation="horizontal")
plt.legend()
```

```python
# !!!BENDE LABELLAR ÇIKMIYOR NEDEN OLDUĞUNU ANLAMADIM!!!
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]

plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis("equal")  # bar chart düzgün görünmezse bunu yaptığımızda düzeliyor
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%")  # radius ile bar chart'ın boyutunu artırdık  # autopct ile içine yazı yazdırdık
plt.show()  # Üstte çıkan yazıları kapattık
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", explode=[0,0.2,0,0,0], startangle=45)  # explode ile dışarı çıkık görünmesini istediğimiz yeri belirttik  # startangle ile 45 derece döndürdük
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
