```python
cars = sns.load_dataset("mpg")
```

```python
type(cars)
```

```python
cars.dropna(inplace=True)
cars.shape
```

```python
cars.head()
```

```python
sns.relplot(x="model_year", y="mpg", col="origin", hue="cylinders", data=cars);  # veri olarak üstteki cars verisini kullan ve seaborn özelliği olan relplot çizimini yap dedim.
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[diamonds.cut.isin(["Premium", "Good"]) & diamonds.color.isin(["D", "F", "J"])].sample(n=100, random_state=22)
diamonds.shape
```

```python
diamonds
```

```python
sns.set_style("dark")  # Stilimi seçtim
```

```python
sns.scatterplot(diamonds.carat, diamonds.price); # ; yerine alt satıra plt.show() diyebilirdim
```

```python
sns.scatterplot(x="carat", y="price", data=diamonds);  # Üstteki ile aynı grafiği verir
```

```python
sns.scatterplot(x="x", y="y", data=diamonds);
```

```python
sns.set_style("darkgrid")
```

```python
sns.color_palette()
```

```python
sns.scatterplot(x="carat", y="price", style="cut", hue="cut", data=diamonds);  # palette ekleyemedim sanırım versiyon hatası veriyor
```

```python
sns.scatterplot(x="carat", y="price", hue="depth", data=diamonds);
```

```python
sns.scatterplot(x="carat", y="price", style="cut", s=100, data=diamonds);  # s control the size
```

```python
sns.scatterplot(x="carat", y="price", style="cut", s=10, data=diamonds); 
```

```python
sns.scatterplot(x="carat", y="price", hue="cut", style="color", data=diamonds, s=100);  # Bende daha düzensiz oldu
```

```python
sns.scatterplot(x="carat", y="price", hue="depth", data=diamonds, s=100, alpha=0.3);
```

```python
sns.scatterplot(x="carat", y="price", hue="depth", data=diamonds, s=100, alpha=0.8);
```

```python
sns.scatterplot(x="carat", y="price", hue="depth", data=diamonds, s=200, marker="*", edgecolor="black", alpha=0.8);
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)  # dropna() function return Index without NA/NaN values. All the missing values are removed and a new object is returned which does not have any NaN values present in it.
```

```python
penguins.shape
```

```python
sns.set_style("darkgrid")
```

```python
sns.histplot(penguins.bill_depth_mm);
```

```python
sns.histplot(y = penguins.bill_depth_mm);  # üsttekini horizontal hale getirdik
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins);  # alternatif başka bir yol
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, kde=True);
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, bins=20);
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, bins=[14, 18, 20]);  # bin için direkt lokasyon belirttik
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, binwidth=2, binrange=(14,20));  # binwidht=2 diyerek 2'şer aralıklı olmasını sağladık
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, stat="density");  # y ekseninin default'u count'tur ama istersek bunu değiştirebiliriz. Alttaki görsel density(yoğunluk)'a bağlı değişim grafiği.
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, stat="probability");  # Alttaki görsel olasılıksal dağılımını gösteriyor
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, stat="probability", fill=False, element='step', cumulative=True); 
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, hue="species")
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, hue="species", element="poly")  # poligram çizdirdik
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, hue="species", multiple="stack")  # multiple'nin defaultu layer
```

```python
sns.histplot(x = "bill_depth_mm", data=penguins, hue="species", multiple="fill") 
```

```python
sns.histplot(x = "bill_depth_mm", y = "bill_depth_mm", data=penguins, cbar=True)
```

```python
sns.histplot(x ="bill_depth_mm", hue="sex", multiple="dodge", data=penguins, shrink=0.8);
```

```python
# Direkt tüm renkleri bir renk yapmak istiyorsam color=... derim. Eğer hepsine ayrı renk vermek istersem palette=... derim
```

```python
sns.histplot(x="species", hue="sex", multiple="dodge", data=penguins, shrink=0.8, palette="bone");
```
